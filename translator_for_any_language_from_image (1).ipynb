And finally now we translate word and showing result

```python
# Check the word and then translate
print(f"The original word is: {word}")

# check the language first
language = translator.detect(word)
print(f"The language of the word is: {language}")

# Now, convert this word into urdu
text_ur = translator.translate(word, src=language.lang, dest='ur').text
print(f"The translated word is: {text_ur}")
```

And finally now we translate 2nd word and showing result

```python
# Check the word and then translate
print(f"The original word is: {word2}")

# check the language first
language2 = translator.detect(word2)
print(f"The language of the word is: {language2}")

# Now, convert this word into urdu
text_ur2 = translator.translate(word2, src=language2.lang, dest='ur').text
print(f"The translated word is: {text_ur2}")
```
